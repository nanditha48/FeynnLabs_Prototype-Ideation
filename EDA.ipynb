```python
#Data Cleaning
#Creating a copy for base data
dt_new = dt.copy()
```

```python
dt_new.TotalCharges = pd.to_numeric(dt_new.TotalCharges, errors='coerce')
dt_new.isnull().sum()
```

```python
#11 missing values present 
dt_new.loc[dt_new ['TotalCharges'].isnull() == True]
```

```python
#Missing Values
#Removing missing values 
dt_new.dropna(how = 'any', inplace = True)

#telco_data.fillna(0)
```

```python
#Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...
# Get the max tenure
print(dt_new['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

dt_new['tenure_group'] = pd.cut(dt_new.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
dt_new['tenure_group'].value_counts()
```

```python
#remove the columns not needed
#drop column customerID and tenure
dt_new.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
dt_new.head()
```

```python
#Data Exploration
#Univariate Analysis
for i, predictor in enumerate(dt_new.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=dt_new, x=predictor, hue='Churn', color="r")
    plt.savefig('test.png', bbox_inches='tight')
```

```python
#Label Encoding for target variables 
dt_new['Churn'] = np.where(dt_new.Churn == 'Yes',1,0)
```

```python
dt_new.head()
```

```python
dt_new_dummies = pd.get_dummies(dt_new)
dt_new_dummies.head()
```

```python
sns.lmplot(data=dt_new_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
plt.savefig('test1.png', bbox_inches='tight')
```

```python
#Churn by Monthly Charges and Total Charges
Mth = sns.kdeplot(dt_new_dummies.MonthlyCharges[(dt_new_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(dt_new_dummies.MonthlyCharges[(dt_new_dummies["Churn"] == 1) ],
                ax =Mth, color="Green", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
##CHURN IS HIGH WHEN MONTHLY CHARGES ARE HIGH
```

```python
Tot = sns.kdeplot(dt_new_dummies.TotalCharges[(dt_new_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(dt_new_dummies.TotalCharges[(dt_new_dummies["Churn"] == 1) ],
                ax =Tot, color="Green", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

```python
#Build a corelation of all predictors with 'Churn'
plt.figure(figsize=(20,8))
dt_new_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar', color="red")
plt.savefig('test2.png', bbox_inches='tight')
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```
